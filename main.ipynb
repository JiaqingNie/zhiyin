```python
from a_star_player import *
from puzzle import *
from utils import *
import matplotlib.pyplot as plt
from IPython import display
import cv2
```

```python
SIZE = 3
image = cv2.imread('onlyBecause.jpg')
shape = image.shape
height = shape[0] // SIZE
width = shape[1] // SIZE
```

```python
def get_assembled_image(img, puzzle):
    image = np.zeros((height * SIZE, width * SIZE, 3), dtype=np.uint8)
    for k, num in enumerate(puzzle):
        a, b = k // SIZE, k % SIZE
        i, j = num // SIZE, num % SIZE
        image[height * a: height * (a + 1), width * b: width * (b + 1)] = img[height * i: height * (i + 1), width * j: width * (j + 1)]
    return image
```

```python
env = Puzzle(3)
player = AStarPlayer(env)
res = player.solve()

for r in res:
    if r == 'up':
        env.move_up()
    elif r == 'down':
        env.move_down()
    elif r == 'left':
        env.move_left()
    elif r == 'right':
        env.move_right()
    plt.axis('off')
    plt.imshow(get_assembled_image(image, env.snapshot()))
    
    display.clear_output(wait=True)
    display.display(plt.gcf())
```
